```python
df = pd.read_csv('BitcoinHeistData.csv')
```

```python

from sklearn.preprocessing import OneHotEncoder,LabelEncoder
# encode label using OneHotEncoder

len(df)
# df.drop(['address'],axis=1,inplace=True)
df
```

```python
from sklearn.compose import make_column_transformer
lb=LabelEncoder()
df['address']=lb.fit_transform(df['address'])
df['label']=lb.fit_transform(df['label'])
```

```python
df
```

```python
Y = df['label']
X = df.drop('label',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.5,shuffle=False)
```

```python

from joblib import Parallel, delayed,dump
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model_rf = RandomForestClassifier(max_depth=2,random_state=40)
model_dt = DecisionTreeClassifier(splitter="random",max_depth=2,random_state=2,max_features="auto")
models = [model_rf, model_dt]
```

```python
import pickle
import gzip


def train_model(X,Y,model,count):
    print('Execution model number',count)
    output = model.fit(X,Y)
    with gzip.open(f"./{count}.pkl","wb") as ofp:
        pickle.dump(output,ofp)
    return output
```

#### Train Parallel

```python
%%time

result = Parallel(n_jobs=2)(delayed(train_model)(X_train, Y_train, models[i],i) for i in range(len(models)))
```

##### Train Sequentially

```python
%%time
s = time.time()
results = [train_model(X,Y,model,i) for i, model in enumerate(models)]
print(time.time()-s)
```

```python
def test_model(X_test,y_test,model,model_count):
  filename = f'{str(model_count)}.pkl'
  with gzip.open(filename, 'rb') as ifp:
    loaded_model=pickle.load(ifp)
    res=loaded_model.predict(X_test)
  return res
```

```python
%%time

result = Parallel(n_jobs=2)(delayed(test_model)(X_test, Y_test, result[i],i) for i in range(len(models)))
```

```python
%%time
result =[test_model(X_test,Y_test,model,i) for i, model in enumerate(models)]
```

```python
result_df = pd.DataFrame({'RF':result[0],'DT':result[1]})
```
