```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df
```

```python
print(df.Pclass.unique())
```

```python
from matplotlib import pyplot as plt
```

```python
# function to make bar plot of number of male, female and mean fair
values = {
    'male':df['Sex'].tolist().count('male'),
    'female':df['Sex'].tolist().count('female'),
    'mean_fair':df['Fare'].mean()
}
values
```

```python
# bar plot
fig,ax = plt.subplots(figsize=(20,10))
plt.bar(list(values.keys()),list(values.values()),color='maroon')
plt.xlabel("Men and Women Mean fair")
plt.ylabel("Values")
datas = list(values.values())
for i,data in enumerate(datas):
    plt.text(x=i,y=data+1,s=f"{int(data)}",fontdict=dict(fontsize=10))
plt.tight_layout()
    
plt.show()
```
