##### Display the mean and standard deviation of “Glucose” in the following format:
#### Mean (std) = mean_value (Std_Vale)

```python
# display mean and standard deviation of a column
def mean_value(df, column):
    return df[column].mean(),df[column].std()

mean,std =mean_value(df,'GLUCOSE')
print(f"Mean of Glucose = {mean}\nStandard Deviation of Glucose = {std}")
```

##### Write a function to find the median value of “BMI” for both 0 and 1 outcome classes 

```python
#function to find median value of BMI for diabetic patients as well as non diabetic patients
def median_value(df, column):
    diabetic = df[df['OUTCOME'] == 1]
    non_diabetic = df[df['OUTCOME'] == 0]
    diabetic_median = diabetic[column].median()
    non_diabetic_median = non_diabetic[column].median()
    return diabetic_median, non_diabetic_median

diabetic_median, non_diabetic_median = median_value(df,'BMI')
print(f"Median of BMI for diabetic patients = {diabetic_median}\nMedian of BMI for non diabetic patients = {non_diabetic_median}")
```

##### Create a new feature named “Age_Category” and add it as last column based on following 
###### conditions:
> If Age >= 50, Age_Category = Old

> If 50> Age >21, Age_Category = Middle_Young

> If Age == 21, Age_Category = Young

```python
#function to add Age_Category column to the dataframe based on the age value
def create_age_category(df):
    df['AGE_CATEGORY'] = df['AGE'].apply(lambda x: 'Young' if x == 21 else 'Middle_Young' if x >21 and x<50 else 'Old')
    return df

df = create_age_category(df)
df
```

##### The zero (0) values presents in "Glucose", "BloodPressure", "SkinThickness" and "Insulin" attributes replace with ‘NaN’ and then fill all NaN entries with corresponding attribute’s mean value.


```python
# replace 0 values in column with NaN
from typing import List
import numpy as np
def replace_zero(df, columns:List[str]):
    for column in columns:
        df[column] = df[column].replace(0, np.NaN)
    return df
df = replace_zero(df,['BMI','GLUCOSE','BLOODPRESSURE','SKINTHICKNESS','INSULIN'])
#fill nan with mean values of the column
def replace_avg(df,columns: List[str]):
    for column in columns:
        df[column].replace(np.nan,df[column].mean(),inplace=True)
    return df
df = replace_avg(df,['BMI','GLUCOSE','BLOODPRESSURE','SKINTHICKNESS','INSULIN'])
df
```
